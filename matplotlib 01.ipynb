```python
import matplotlib.pyplot as ka #untuk mengimport module
import numpy as ii
```

```python
#data Sample
x = [5,6,7,8] #data sample x
y = [5,8,7,6] #data sample y

fig, ax = ka.subplots() #memvisualisasikan data x dan y dengan membuat figure dan axes
ax.plot(x,y)
```

```python
x = [5,6,7,8] #data sample x
y = [6,8,5,7] #data sample Y

ka.plot(x,y) #memulai ploting data x dan y
```

```python
fig = ka.figure() #membuat suatu figure tanpa disertai axes
```

```python
fig, ax = ka.subplots() #membuat suatu figure disertai axes
```

```python
fig, ax = ka.subplots(2,3) #membuat suatu figure yang disertai axes 2x3 grid
```

```python
x = ii.linspace(0, 2, 100) #mengenerate sejumla deret bilangan
x
```

```python
fig, ax = ka.subplots() # membuat figure yang disertai dengan axes

#memploting 3 type data yang ada pada axes
ax.plot (x, x, label = 'linear')
ax.plot (x, x**2, label = 'quadratic')
ax.plot (x, x**3, label = 'cubic')

ax.set_xlabel('x label') # menyertakan x-label yang terdapat pada axes
ax.set_ylabel('y label') # menyertakan y-label yang terdapat pada axes
ax.set_title('sample plot') # menyertakan title yang terdapat pada axes
ax.legend() #menyertakan legend
```
