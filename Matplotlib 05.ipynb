```python
import matplotlib
import matplotlib.pyplot as ka #untuk mengimport module
import numpy as ii

print(matplotlib.__version__)
print(ii.__version__)
```

```python
data = [25, 45, 55, 125, 225]
kategori = ['G', 'H', 'I', 'J', 'K']
```

```python
#membuat simple Bar Plot
ka.bar(kategori, data)
ka.grid()
ka.xlabel('kategori')
ka.ylabel('jumlah')
ka.title('contoh Bar Plot')

ka.show()
```

```python

data = [25, 45, 55, 125, 225]
kategori = ['G', 'H', 'I', 'J', 'K']
```

```python
#mengatur grid dan color
ka.bar(kategori, 
       data,
      color= 'red',
      alpha= 0.25)

ka.grid(linestyle ='--',
       linewidth=2,
       axis='y',
       alpha =0.50)
ka.xlabel('kategori')
ka.ylabel('jumlah')
ka.title('contoh Bar Plot')

ka.show()
```

```python
data1 = [25, 85, 75, 40, 60]
data2= [40, 35, 20, 55, 10]
kategori = ['G', 'H', 'I', 'J', 'K']
```

```python
#membuat stacked bar plot
ka.bar(kategori, data1, label='data1')
ka.bar(kategori, data2, label='data2', bottom=data1)

ka.legend()
ka.grid(linestyle ='--',
       linewidth=1,
       axis='y',
       alpha =0.75)
ka.xlabel('kategori')
ka.ylabel('jumlah')
ka.title('contoh Bar Plot')

ka.show()
```

```python
data1 = [25, 85, 75, 40, 60]
data2= [40, 35, 20, 55, 10]
kategori = ['G', 'H', 'I', 'J', 'K']

x = ii.arange(len(kategori))
x
```

```python
width =0.25

ka.bar(x-width/2, data1, width, label='data1')
ka.bar(x+width/2, data2, width, label='data2')


ka.xticks(x, kategori)
ka.grid(linestyle ='--',
       linewidth=1,
       axis='y',
       alpha =0.75)

ka.xlabel('kategori')
ka.ylabel('jumlah')
ka.title('contoh Bar Plot')

ka.legend()
ka.show()
```
