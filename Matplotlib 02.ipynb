```python
import matplotlib
import matplotlib.pyplot as ka #untuk mengimport module
import numpy as ii
```

```python
#plotting sederhana
ka.plot([2,5,7,11])
ka.ylabel('sumbu Y')
ka.show()
```

```python
#multi plot satu fungsi pemanggilan
ka.plot([2,3,4,7],[5,4,2,16])
ka.show()
```

```python
#pengaturan format pada plot
ka.plot([1,2,3,4],[1,4,9,16],'ro')
ka.axis([0,6,0,20])
ka.show()
```

```python
t = ii.arange(0.,5.,0.2)
t 
```

```python
ka.plot(t,t,'r--')
ka.plot(t,t**2,'bs')
ka.plot(t,t**3,'g^')
ka.show()
```

```python
data = {'o':ii.arange(50),
       'p':ii.random.randint(0,50,50),
       'q':ii.random.randn(50)}
data['p'] = data['o']+10*ii.random.randn(50)
data['q'] = ii.abs(data['q'])*100

data
```

```python
ka.scatter('o','p',data=data)
ka.xlabel('entry o')
ka.ylabel('entry p')
ka.show()
```

```python
names = ['group_o','group_p','group_q']
values = [1,10,100]
```

```python
ka.plot(names,values)
ka.show()
```

```python
ka.scatter(names,values)
ka.show()
```

```python
ka.bar(names,values)
ka.show()
```
