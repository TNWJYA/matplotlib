```python
import matplotlib
import matplotlib.pyplot as ka #untuk mengimport module


print(matplotlib.__version__)
```

```python
kategori = ['G', 'H', 'I', 'J', 'K']
data = [25, 45, 55, 125, 225]
```

```python
ka.pie(data, 
       labels=kategori,
      autopct= '%1.1f%%',
     startangle= 90)


ka.title('contoh pie Plot')

ka.show()
```

```python
kategori = ['G', 'H', 'I', 'J', 'K']
data = [25, 45, 55, 125, 225]
warna = ['pink', 'cyan','#e03364', 'yellowgreen', 'skyblue']
```

```python
ka.pie(data, 
       labels=kategori,
       colors= warna,
      autopct= '%1.1f%%',
     startangle= 90)

ka.legend()
ka.title('contoh pie Plot')

ka.show()
```

```python
kategori = ['G', 'H', 'I', 'J', 'K']
data = [215, 130,245, 215,225]
warna = ['pink', 'cyan','#e03364', 'yellowgreen', 'skyblue']
explode_var =[0.,0., 0.2, 0., 0.08]
```

```python
ka.pie(data, 
       labels=kategori,
       colors= warna,
       explode=explode_var,
      autopct= '%1.1f%%',
     startangle= 90)


ka.title('contoh pie Plot')

ka.show()
```
