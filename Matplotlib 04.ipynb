```python
import matplotlib
import matplotlib.pyplot as ka #untuk mengimport module
import numpy as ii

print(matplotlib.__version__)
print(ii.__version__)
```

```python
mu, sigma = 100, 15 
x = mu + sigma * ii.random.randn(10000)
x
```

```python
x.shape
```

```python
ka.hist(x,
       bins = 50,
       facecolor='g',
        alpha=0.75)
ka.xlabel('sumbu X')
ka.ylabel('Sumbu Y')
ka.title('Contoh Histogram')

ka.text(45,500,r'$\mu=100,\ \sigma=15$')
ka.grid()

ka.show()

```

```python
fig, ax = ka.subplots()

ax.hist(x,
       bins = 50,
       facecolor='r',
        alpha=0.75)
ax.set_xlabel('sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45,500,r'$\mu=100,\ \sigma=15$')
ax.grid()

ka.show()
```
