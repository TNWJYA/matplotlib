```python
import matplotlib
import matplotlib.pyplot as ka
import numpy as ii

print(matplotlib.__version__)
print(ii.__version__)
```

```python
fig = ka.figure()
ax = ka.axes(projection='3d')

ka.show()
```

```python
z = ii.linspace(0, 1, 100)
x = z * ii.sin(20 * z)
y = z * ii.cos(20 * z)
```

```python

fig = ka.figure()
ax = ka.axes(projection='3d')

ax.plot(x, y, z, color='red')

ax.set(title='Contoh 3D Line Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

ka.show()
```

```python
fig = ka.figure()
ax = ka.axes(projection='3d')

ax.scatter(x, y, z, color='green')

ax.set(title='Contoh 3D Scatter Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

ka.show()
```
