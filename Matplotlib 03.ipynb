```python
import matplotlib
import matplotlib.pyplot as ka #untuk mengimport module
import numpy as ii

print(matplotlib.__version__)
print(ii.__version__)
```

```python
x = ii.arange(0.0,2.0,0.01)
x
```

```python
s = ii.sin(2*ii.pi*x) 
s
```

```python
fig, ax = ka.subplots() #00 style
ax.plot(x, s)
ax.set(xlabel='nilai x',
      ylabel='niali sin (X)',
      title='Visualisasi nilai sin')

ax.grid()

ka.show()
```

```python
x1 = ii.linspace(0.0,5.0,100)
x2 = ii.linspace(0.0, 2.0,100)

y1 = ii.cos(2*ii.pi*x1)
y2 = ii.cos(2*ii.pi*x2)
```

```python
x1
```

```python
fig, (ax1,ax2) = ka.subplots(2,1)
fig.suptitle('Multiple Subsplots')

ax1.plot(x1, y1,'ro-')
ax1.set_ylabel('nilai cos(x1)')

ax2.plot(x2, y2,'g.-')
ax2.set_ylabel('nilai cos(x2)')

ax2.set_xlabel('Nilai x')
```

```python
ka.subplot(2,1,1)
ka.plot(x1,y1, 'ro-')
ka.title('multiple subsplots')
ka.ylabel('niali $cos(x_1)$')

ka.subplot(2,1,2)
ka.plot(x2,y2, 'g.-')
ka.title('niali $cos(x_2)$')
ka.ylabel('niali $x$')

ka.show()
```
